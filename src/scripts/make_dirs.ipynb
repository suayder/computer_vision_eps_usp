```python
CLASS_NAME = 'pen'
NUM_OBJECTS = 3
NUM_BACKGROUND = 3
```

```python
import os

path = f'stage/{CLASS_NAME}/'

csv_description='env,bg,description\n'
for enviroment in list(['indoor', 'outdoor']):
  for bg in range(NUM_BACKGROUND):
    csv_description+=f'{enviroment},bg{bg},\n'
    for lighting in list(['day', 'night']):
      for obj in range(NUM_OBJECTS):
        full = os.path.join(path, enviroment, lighting, f'object_{obj}',f'bg{bg}')
        os.makedirs(full, exist_ok=True)
with open(f'{path}describe.csv','w') as f:
  f.write(csv_description)
print(f'created folder for class in {path}')
```
